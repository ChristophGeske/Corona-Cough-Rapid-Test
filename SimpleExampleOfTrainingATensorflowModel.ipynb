```python
# An introduction to the fundamentals of programming such a neural network can be learned here: https://www.youtube.com/watch?v=tPYj3fFJGjk

import tensorflow as tf
import numpy as np
from tensorflow import keras,lite
from google.colab import drive
# You can put a '#' infront of this line to run the script without connecting to google drive but than you wont have a location to store the model that is generated.
# drive.mount('/content/gdrive')

x_data = np.array([-1.0,0.0,1.0,2.0,3.0,4.0],dtype=float)
# with y=2x-1 as the function the model needs to learn
y_labels = np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=float)

# Layers, such as tf.keras.layers.Dense, have parameters that are learned during training.
model = keras.Sequential([
                          keras.layers.Dense(units=1,input_shape=[1]), # Input layer with one neuron. Accepting the input for x.
                          keras.layers.Dense(units=1,input_shape=[1])  # Output layer. returning the result in y
                          ])
# An optimizer and the loss function are also known as hyper parameters and can be replaced by other methods to improve the results.
# The optimizer is the algorithem that performs the gradient decent on the network.
# TODO: The accuracy in this example gets stuck at 0.1667 but I am not sure why?
model.compile(optimizer='sgd',loss='mean_squared_error', metrics=['accuracy'])

# fit means training the model. x_data herby represents the training data points, y_labels represents the training labels, epochs is another hyper-parameter that can be changed and describes how often a small training step is repeated
# In the output using 500 epochs you notice that after 390 epochs the loss doesn't change anymore so 400 epochs should be fine also making the training faster.
model.fit(x_data,y_labels,epochs=400)
print('When setting x=10 and giving it to the model as an example, remember the function the model should have learned by now is y=2x-1, the model returns the following result:')
print(model.predict([10]))

# Saves the model to a HDF5 file but not the weights! See https://stackoverflow.com/questions/62832159/how-to-save-full-tensorflow-model-from-google-colab
# The '.h5' extension indicates that the model should be saved to HDF5. 
keras_file = "linear.h5"
# just Download .h5 file by clicking on the 'Files' Tab on the left and Download linear.h5 OR If you prefere: How to save to Google drive followed this explaination: https://medium.com/@ml_kid/how-to-save-our-model-to-google-drive-and-reuse-it-2c1028058cb2 
tf.keras.models.save_model(model,keras_file)

converter = lite.TFLiteConverter.from_keras_model(model)
tfmodel = converter.convert()

# Download .tflite file by clicking on the 'Files' Tab on the left and Download linear.tflite probably in the /content/... folder if you cennected your google drive
open("linear.tflite","wb").write(tfmodel)
```
